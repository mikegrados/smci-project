## Introduction

**Background**

- The democratic world is concerned with the decline in voter turnout as it compromises political representation and political equality (Kostelka et al., 2024)
- The empirical research on this relevant issue gained momentum in the 2000's (2010's)
- Political scientists pointed out the effects of compulsory voting (CV) in citizen’s participation:
  - Since CV raises turnout through an artificial process, some scholars have argued that CV systems *differ in important ways* from non-compulsory systems with high levels of political participation (Ferwerda, 2014)
  - Despite its magnitude, the effect of CV *disappears quickly* after voting is no longer compulsory (Bechtel et al., 2018)
  - There is *no evidence* that CV instils voting habits (Dunaiski, 2021)
  - Kostelka et al. (2024) focused on *democratic states* in 1945-2017 and used FE-regression model to discover the importance of *sanctions for abstention* on the CV's effect on turnout

<center><b>Hypothesis</b></center>

<br/>

<center><em>The introduction of CV increases citizen participation in elections only in the first post-treatment period, followed by a modest decline of turnout in the subsequent periods</em></center>

<center>(one tailed test)</center>

 <center><b>Research question</b></center>

 <br/>

 <center><em>By how much, on average, does the CV introduction increase the turnout?</em></center>

 <center>(ATT estimation)</center>

**Design specification**

- Subjects are *countries*
- The outcome variable is *Turnout*
- The treatment variable is *adoption of CV*
- We use the TWFE model to estimate the causal effect and support it with a staggered difference-in-differences approach (de Chaisemartin &  D'Haultfoeuille, 2020)

**Data specification**

- Data type: panel data for countries (presidential and legislative elections)
- Study period: from 1945 and 2023
- Source: [International IDEA website](https://www.idea.int/data-tools/data/voter-turnout-database)
- Number of records: 2464
- Time imbalance: Yes

## Preparatory material

**I. Two way fixed effects**

From Wooldridge, J. (2021), TWFE estimator includes:
  - Unit fixed effects to remove unit-specific time averages
  - Time fixed effects to remove secular changes in the economic environment that have the same effect on all units



*TWFE assumptions*

- Parallel Trends
- No anticipation
- Treatment responsiveness is homogeneous across treated subjects
- Conditional on treatment group status, treatment exposure is homogeneous across subjects
- ATTs are homogeneous over time
- All treated subjects are treated at the same time

**II. Staggered Difference-in-differences**

From (de Chaisemartin & D'Haultfoeuille, 2020):

$$\textrm{DID}_M = \sum_{t=2}^{T} \left( \frac{N_{1,0,t}}{N_S}\textrm{DID}_{+,t} + \frac{N_{0,1,t}}{N_S}\textrm{DID}_{-,t} \right),$$

where $N_S = \sum_{(g,t):t\geq 2, D_{g,t} \neq D_{g,t-1}} N_{g,t}$,
$\textrm{DID}_{+,t}$ accounts for the treated,
and $\textrm{DID}_{-,t}$ accounts for the quiters




*Staggered DiD assumptions*
- Balanced Panel of Groups
- Sharp Design
- Common Trends for $Y(0)$ and $Y(1)$
- Strong Exogeneity for $Y(0)$ and $Y(1)$
- Existence of "Stable" Groups
- Mean Independence between a Group's Outcome and Other Groups Treatments

**Not violated**
- *No anticipation:* treatment takes effect on the first elections after CV is introduced
- *Conditional on treatment group status, treatment exposure is homogeneous across subjects:* CV is either adopted or not

**Yet some assumptions do not hold (TWFE results are not robust)**
- *Treatment responsiveness is homogeneous across treated subjects:* there may be some countries whose turnout was high enough before CV adoption; therefore, change is not considerable
- *ATTs are homogeneous over time:* the turnout in the first elections after the adoption of CV (should) reach a peak and decline afterwards
- *All treated subjects are treated at the same time:* time imbalance
